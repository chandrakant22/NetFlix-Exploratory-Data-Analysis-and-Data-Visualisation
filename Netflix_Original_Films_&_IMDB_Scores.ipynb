```python
#Print only movies which IMDB Score is Greater than 7 
score_8 = ds[ds['IMDB Score']>=8]
score_8
```

```python
# top 10 movies in imdb rating more than 7
sc=score_8[::-1]
sc.head(10)
```

```python
# average runtime in dataset
np.average(ds['Runtime'])
```

```python
# Print only movies which IMDB Score is Greater than 7 and runtime less than 60
runtimels60 = score_8[score_8['Runtime']<=60]
runtimels60
```

```python
# Print only name of movies and Language which IMDB Score is Greater than 7 and runtime less than 60
runtimels60 = score_8[score_8['Runtime']<=60]
runtimels60['Title']+" "+runtimels60['Language']
```

```python
# Only 3 movies which has ratings over 8 are not in english language
nh=score_8[score_8['Language']!='English']
nh[:3]
```

```python
# IMDB Movie ratings less than 5 and their languages
score_ls5 = ds[ds['IMDB Score']<5]
score_ls5['Title']+" | "+score_ls5['Language']
```

```python
# Top 10 Content Language Dataset
all_lang = ds['Language'].value_counts()
```

```python
all_lang[:10]
```

```python
# Top 10 Content Language Dataset Visualisation
all_lang[:10].plot(kind='barh')
```
